Q2. MOST SELLING PRODUCTS

```python
# BY   QUANTITY
data.head(5)
```

```python
# Most selling products by quantity in descending order
data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
# 
most_quntity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
most_quntity
```

```python
# Checking the data type of most_quntity
type(most_quntity) 
```

```python
# Set most_quntity as a DataFrame
most_quntity = most_quntity.reset_index()
most_quntity   
```

```python
#Checking the datatype now
type(most_quntity)
```

```python
# plotting the most selling products by quantity
plt.figure(figsize=(9,4))
plt.bar(most_quntity['Product'], most_quntity['Quantity'], color = ['red', 'black', 'green', 'yellow', 'cyan'], width=0.5)
plt.title('Most Selling Products by Quantity')
plt.xlabel('Products')
plt.ylabel('Quantity Sold')
```

```python
# BY REVENUE
data.head(5)
```

```python
# adding a new column 'Revenue' to the dataset
data['Revenue'] = data['Quantity'] * data['Price']
data
```

```python
# most selling products by revenue in descending order
data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
#
Most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
Most_revenue
```

```python
# 
Most_revenue = Most_revenue.reset_index()
Most_revenue
```

```python
#
plt.figure(figsize=(9,4))
plt.bar(Most_revenue['Product'], Most_revenue['Revenue'], color = ['blue', 'orange', 'green', 'red', 'purple'], width=0.5)
plt.title('Most Selling Products by Revenue')
plt.xlabel('Products')
plt.ylabel('Revenue Generated')
```
